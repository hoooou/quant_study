```python
#打印十年行情
import mplfinance as mpf
import pandas as pd
stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol="000001", period="daily", start_date="20170301", end_date='20240528', adjust="")
stock_zh_a_hist_df["日期"] = pd.to_datetime(stock_zh_a_hist_df["日期"])
stock_zh_a_hist_df = stock_zh_a_hist_df.set_index(["日期"])
stock_zh_a_hist_df['Open'] = stock_zh_a_hist_df['开盘']
stock_zh_a_hist_df['High'] = stock_zh_a_hist_df['最高']
stock_zh_a_hist_df['Low'] = stock_zh_a_hist_df['最低']
stock_zh_a_hist_df['Close'] = stock_zh_a_hist_df['收盘']
stock_zh_a_hist_df['Volume'] = stock_zh_a_hist_df['成交量']
stock_zh_a_hist_df = stock_zh_a_hist_df["2017-04-01": "2020-04-29"]
# print(stock_zh_a_hist_df)
mpf.plot(stock_zh_a_hist_df, type="candle", mav=(3, 6, 9), volume=True, show_nontrading=False)
```
